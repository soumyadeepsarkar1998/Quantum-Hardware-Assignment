```python
import numpy as np
import math
```

```python
import matplotlib.pyplot as plt
```

```python
from scipy import linalg
```

```python
pi= np.pi
sigmaz = np.array([ [1,0] , [0,-1] ])
sigmax = np.array( [ [0,1] , [1,0]])
sigmay = np.array([[0,-1j] , [1j,0] ])
identity = np.array( [ [1,0],[0,1] ])
```

```python
Omega = 2*pi
w = 2*pi
deltat = 0.01
t = np.linspace(0, 2, num=201)
init = np.array([1,0])
phi = np.array([init])
psi = np.array([init])
```

```python
for i in t:
    phi=np.append(phi, [np.dot(linalg.expm((1j)*(deltat/2)*(w*sigmaz+Omega*math.cos(w*i)*sigmax-Omega*math.sin(w*i)*sigmay)), phi[-1])] , axis=0)
    psi=np.append(psi, [np.dot(linalg.expm((1j)*(deltat/2)*(w*sigmaz+2*Omega*math.cos(w*i)*sigmax)), psi[-1])] , axis=0)
```

```python
Pphizero = np.array([ abs(x[0])*abs(x[0]) for x in phi])
Ppsizero = np.array([ abs(y[0])*abs(y[0]) for y in psi])
```

```python
t=np.append(t,2.01)
```

```python
plt.plot(t, Pphizero, label="part a")
plt.plot(t, Ppsizero, label ="part b")
plt.show()
```
